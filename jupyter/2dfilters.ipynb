# 2D Filters



```clojure
(ns voiceless-drylands
    (:require
    [opencv4.utils :as u]
    [opencv4.core :refer :all]))
```

```clojure
(def neko 
  (-> "resources/images/cat.jpg" 
      imread  
      (u/resize-by 0.2)))
```

## prewitt-horizontal

```clojure
(defn apply-my-filter 
  ([ mat my-filter] 
   (apply-my-filter mat my-filter false))
  ([ mat my-filter invert]
  (->
  mat 
  clone
  (cvt-color! COLOR_RGB2GRAY)
  (filter-2-d! -1 (u/matrix-to-mat my-filter))
  ((fn [x] (if invert (bitwise-not! x) x)))
  (u/>>> ))))

(apply-my-filter neko
     [ [-1 -1 -1]
       [0 0 0]
       [1 1 1]] true)
```

## prewitt-vertical

```clojure
(->
  neko 
  clone
  (cvt-color! COLOR_RGB2GRAY)
  (filter-2-d! -1 (u/matrix-to-mat
    [ [-1 0 1]
      [-1 0 1]
      [-1 0 1]]))
  (u/>>> ))
```

## sobel vertical-matrix

```clojure
(->
  (imread "resources/images/cat.jpg" IMREAD_REDUCED_GRAYSCALE_4)
  (filter-2-d! -1 (u/matrix-to-mat
    [ [-1 0 1]
      [-2 0 2]
      [-1 0 1]]))
  (u/>>>))
```
