## Convert a cat to small picture, in black and white

```clojure
(-> "https://raw.githubusercontent.com/hellonico/origami-fun/master/resources/3-kittens.jpg"
    u/mat-from-url
    (u/resize-by 0.2)
    (cvt-color! COLOR_RGB2GRAY)
    (u/>>>))
```
