# lesson: apply blur to a mat

```clojure
(require
 '[opencv4.core :refer :all]
 '[opencv4.utils :as u])

(def neko
  (-> "resources/chapter02/emilie5.jpg"
      (imread)
      (u/resize-by 0.07)))
(u/>>> neko)
```

```clojure
(-> neko  clone (blur! (new-size 3 3)) (u/>>>))
```

```clojure
(->> (range 3 10 2)
     (map #(-> 
	  neko  clone (u/resize-by 0.5) (blur! (new-size % %))))
     (hconcat!)
     (u/>>>))
```

```clojure
(-> neko  clone (gaussian-blur! (new-size 5 5) 0 0) (u/>>>))
```

```clojure
(-> neko 
    clone 
    (bilateral-filter! 9 9 7) 
    (u/>>>))
```

```clojure
(-> neko 
    clone 
    (cvt-color! COLOR_BGR2GRAY) 
    (bilateral-filter! 9 9 7) 
    (canny! 50.0 250.0 3 true) 
    (bitwise-not!)
    (u/>>>))
```

```clojure
(-> neko 
    clone 
    (cvt-color! COLOR_BGR2GRAY) 
    (blur! (new-size 3 3))
    (canny! 50.0 250.0 3 true) 
    (bitwise-not!)
    (u/>>>))
```

```clojure
(-> neko 
    clone 
    (median-blur! 27)
    (u/>>>))
```

```clojure
(-> neko 
    clone 
    (cvt-color! COLOR_BGR2GRAY) 
    (median-blur! 7)
    (canny! 50.0 250.0 3 true) 
    (bitwise-not!)
    (u/>>>))
```
