# Working with masks

```clojure
(ns wooden-winds
   (:require
    [opencv4.core :refer :all]
    [opencv4.utils :as u]))
```

```clojure
(def img "https://images-na.ssl-images-amazon.com/images/I/61qltrwAkIL._SY355_.jpg")
```

```clojure
(def hsv (-> img (u/mat-from-url) (flip! 1) (cvt-color! COLOR_BGR2HSV) ))
```

```clojure
(-> hsv u/>>>)
```

```clojure
(def mask (new-mat)) 
(in-range hsv (new-scalar 0 120 70) (new-scalar 10 255 255) mask)
```

```clojure
(def mask2 (new-mat))
(in-range hsv (new-scalar 170 120 70) (new-scalar 180 255 255) mask2)
```

```clojure
(def mask3 (add! mask mask2))
```

```clojure
(u/>>> mask3)
```
