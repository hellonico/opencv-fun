# Finding and drawing contours
Finding object by their shape is one of the key reason people usually comes to OpenCV. 

```clojure
(ns lush-storm
  (:require
    [opencv4.utils :as u]
    [opencv4.colors.html :as color]
    [opencv4.core :refer :all]))
```

## Finding contour of headphone and drawing rectangles ...

```clojure
(def headphones 
  (-> "resources/chapter03/sonyheadphones.jpg"
      (imread IMREAD_REDUCED_COLOR_4)))

(u/>>> headphones)
```

```clojure
(def mask
  (-> headphones 
      clone 
      (cvt-color! COLOR_BGR2GRAY)
 	  (threshold! 250 255 THRESH_BINARY)
      (median-blur! 7)))

(u/>>> mask)
```

```clojure
(def masked-input (clone mask))
(u/>>> masked-input)
```

```clojure
(def contours 
  (new-arraylist))

(find-contours 
  masked-input
  contours 
  (new-mat) ; mask
  RETR_TREE
  CHAIN_APPROX_NONE)
```
