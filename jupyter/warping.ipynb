```clojure
(ns talented-silence
  (:require
    [opencv4.core :refer :all]
    [opencv4.video :as v]
    [opencv4.colors.rgb :as rgb]
    [opencv4.utils :as u]))
```

### warping

```clojure
(defn warp! [ buffer p1 p2]
  (-> buffer
    (warp-perspective!   
        (get-perspective-transform
           (u/matrix-to-matofpoint2f points1)
           (u/matrix-to-matofpoint2f points2)) 
           (size buffer ))))
```
