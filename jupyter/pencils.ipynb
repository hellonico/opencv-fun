# Pencil Sketch

Here we are using the native opencv function named pencil-sketch. This automatically draw your picture like pencils. This gives a nice non-automatic feeling to your creations.

```clojure
(ns thoughtless-crater
    (:require
    [opencv4.core :refer :all]
    [opencv4.utils :as u]))
```

First we load the image, and then apply the pencil-sketch! function on a clone of the original picture. 

```clojure
(def original
   (-> "resources/images/cat.jpg" (imread) (u/resize-by 0.20)))
(u/>>> original)
```

```clojure
(-> original clone (pencil-sketch! (new-mat)) u/>>>)
```

The result is a bit dark, so to make the work of pencil-sketch easier and with a better result, let's apply a filter first, to highlights edges on the original picture. You'll notice a somewhat blurry effect on the picture.

```clojure
(def edge (-> original clone edge-preserving-filter!))
(u/>>> edge)
```

Then we try again the pencil-sketch function on the filtered image.

```clojure
(-> edge clone (pencil-sketch! (new-mat)) u/>>>)
```
