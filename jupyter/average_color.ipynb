# Average Color

Here we concatenate the original with another image, where the other image has all its pixel set to the average color o the original picture.

```clojure
(ns opencv4.average-color
 (:require
   [opencv4.utils :as u]
   [opencv4.core :refer :all]))
```

```clojure
(defn average-color[input]
  (let [src    (-> input (imread) (u/resize-by 0.20))
        mean_  (-> src u/mat-from (set-to! (mean src))) ]
    (-> src
        (u/mat-from)
        (set-to! (mean src))
        (cons [src])
        (hconcat!)  )))
```

```clojure
(-> 
 "resources/images/cat.jpg" 
 average-color 
 (u/>>>))
```
