```clojure
(defn gamma![_img gamma]
	(let [
		; _img (Imgcodecs/imread img)
		_gamma (/ 1.0 gamma)
		_lut (new-mat 1 256 CV_8UC1)
		]
		(.setTo _lut (new-scalar (double 0)))
		(doseq [i (range 0 256)]
 		(let [v (double-array [ (* 255 (Math/pow (/ i 255) _gamma))] ) ]
 			(.put _lut 0 i v )))
		(lut _img _lut _img)
		_img))
```
