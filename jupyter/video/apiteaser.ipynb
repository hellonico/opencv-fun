# API Teaser

This is a one pager to give you an idea on how to the Clojure wrapper for OpenCV can be used.


```clojure
(ns composed-pine
   (:require
    [opencv4.utils :as u]
    [opencv4.core :refer :all]))
```

```clojure
(defn load-cvt-resize-show[url color width height]
 (-> 
   url
   (u/mat-from-url)
   (cvt-color! color)
   (resize! (new-size width height))
   (u/>>>)))
```

```clojure
(->
   "https://s-media-cache-ak0.pinimg.com/736x/29/9e/56/299e56ab07c75af6407289ecc4ab1dd6--mothers-love-mother-daughters.jpg"
  (load-cvt-resize-show COLOR_RGB2GRAY 150 200))
```

```clojure
(-> 
  "https://i.ytimg.com/vi/p4Jj9QZFJvw/hqdefault.jpg"
  (load-cvt-resize-show COLOR_BGR2BGRA 200 150))
```

```clojure
(-> "https://i.ytimg.com/vi/p4Jj9QZFJvw/hqdefault.jpg"
  (u/mat-from-url)
  (u/resize-by 0.80)
  (u/show))
```
