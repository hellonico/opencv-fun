# Change the exposure (the contrast) of a picture 

Changing the exposure of a picture is rather easy and done simply using the function **convert-to**. 

```clojure
(ns inquisitive-pond
   (:require
     [opencv4.utils :as u]
     [opencv4.core :refer :all]))

(def org 
  (-> "resources/images/cat.jpg" imread (u/resize-by 0.2)))

(u/>>> org)
```

## Making the image darker

```clojure
(def img1
  (->
 org
 clone
 (convert-to! -1 0.4 0)))
(u/>>> img1) 
```

## Making the image brigther

```clojure
(def img2
  (->
  org
  clone
  (convert-to! -1 2 0)))
  (u/>>> img2)
```

```clojure
(def img3 (new-mat))
(add-weighted img1 0.2 img2 0.8 1 img3)
(u/>>> img3)
```
